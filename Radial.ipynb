#### Calculate the water density and velocity in pore

#### GRA: 13 16 20 24 27
#### Phos: 22 22 25 29 33 
#### MoS2: 28 35 39 42 47

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = 'phos'
atoms = '33'
run = '1'
```

```python
# Wrapper function of getfromtxt
def load(mat,atoms,p,run):
    pressure = str(p)
    datafile = np.genfromtxt('%s.%s.%s.%s.radial.txt' %(mat,atoms,pressure,run),delimiter = ' ', 
                                 dtype = int, skip_header = 1, invalid_raise = False)
    # Density are the first 11 columns and the velocity are second 11 columns
    density = datafile[:,:11]
    density = np.mean(density,axis = 0)
    velocity = datafile[:,11:22]
    velocity = np.mean(velocity,axis = 0)
    return density, velocity
```

```python
def plot_den(d):
    for density in d:
        distance = list(range(11))
        plt.plot(distance, density)
    return
```

```python
def plot_v(v):
    for velocity in v:
        distance = list(range(11))
        plt.plot(distance, velocity)
    return
```

```python
den_a, vel_a = load(mat,atoms,100,run)
den_b, vel_b = load(mat,atoms,150,run)
den_c, vel_c = load(mat,atoms,200,run)
den_d, vel_d = load(mat,atoms,250,run)
den_e, vel_e = load(mat,atoms,300,run)
```

```python
density = [den_a,den_b,den_c,den_d,den_e]
velocity = [vel_a,vel_b,vel_c,vel_d,vel_e]
```

```python
plot_den(density)
```

```python
plot_v(velocity)
```
